```python
# mopac energy optimizer 
# amber energy
# pipeline as block

import sys 
sys.path.append('..')

import molsberry.core as core_
import molsberry.modules.generic as generic_
import molsberry.modules.cuby4 as cuby4_
import molsberry.modules.mopac as mopac_
import molsberry.modules.parmed as parmed_
import molsberry.modules.rdkit as rdkit_
from molsberry.core.data.collections import BatchedData

class MyPipe1(core_.Pipeline):
    name = "pipe1"
    display_name = "My Pipe 1"
    def build(self):
        conf_amber = cuby4_.Cuby4AMBERInterfaceConfig(
            home="/home/arian/mambaforge/envs/ASH"
        )
        
        self.add_block(core_.InputBlock(["protein"]), "input")
        self.add_block(core_.OutputBlock(["energy"]), "output")

        self.add_block(parmed_.OpenMMProteinParameterizer(), 'prot_param')
        self.add_connection('input', 'protein', 'prot_param', 'proteins')
        
        self.add_block(cuby4_.Cuby4AMBEREnergyCalculator(interface_config=conf_amber,
                                                          save_output=False),
                                                            'amber_energy')
        self.add_connection('prot_param', 'proteins', 'amber_energy', 'molecules')
        self.add_connection('amber_energy', 'energy', 'output', 'energy')
```
