#### ProDy

```python
from prody import proteins, calcGyradius, view3D
import numpy as np
```

```python
kguD = proteins.parsePDB(path2, bonds=True)
print("loaded:", kguD)

kguD.inferBonds()
print("inferred bonds: ", kguD.numBonds(), "bonds in group.")

print("prot attributes:")
print("radius of gyration:", calcGyradius(kguD))

print("atom attributes:")
print("hetatm atom group:", kguD.hetatm) # TODO

print("viewing 3D structure:")
view3D(kguD)
```

```python
cs = 0
for atom in kguD.iterAtoms():
    if atom.getCharge() != 0.0: print(atom)
    cs += atom.getCharge()
cs
```
