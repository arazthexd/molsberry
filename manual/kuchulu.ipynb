```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.drawOptions.addAtomIndices = True
IPythonConsole.molSize = 700, 500
```

```python
m = Chem.MolFromSmiles("CC(Cl)C(=O)[O-]")
m = Chem.AddHs(m)
AllChem.EmbedMolecule(m)
AllChem.MMFFOptimizeMolecule(m)
m = Chem.RemoveHs(m)
m
```

```python
ionized_idx = 5
mm = Chem.Mol(m)
mm.GetAtomWithIdx(ionized_idx).SetFormalCharge(0)
mm.GetAtomWithIdx(ionized_idx).SetNumExplicitHs(1)
Chem.MolToPDBFile(Chem.AddHs(mm, addCoords=True), "tmp.pdb")
Chem.AddHs(mm, addCoords=True)
```
